```python
# Какая средняя и медианная зарплата по группам вакансий?
b = df2[['job_title',"salary"]]

spisok = []
for i in b.job_title:
    if spisok.count(i) == 0:
        spisok.append(i)
    else:
        continue
sale = []
f = 0
for i in spisok:
    for j in b.job_title:
        if i == j:
            sale.append(b.loc[f,'salary'])
            f += 1
        else:
            f += 1
    df3 = pd.DataFrame((sale),columns=["salary"])
    print(i, "          ", df3.salary.mean(), "          ", df3.salary.median())
    sale.clear()
    f = 0
```

```python
# Какая средняя и медианная зарплата по каждому региону?
b = df2[['company_location',"salary"]]

spisok = []
for i in b.company_location:
    if spisok.count(i) == 0:
        spisok.append(i)
    else:
        continue
sale = []
f = 0
for i in spisok:
    for j in b.company_location:
        if i == j:
            sale.append(b.loc[f,'salary'])
            f += 1
        else:
            f += 1
    df3 = pd.DataFrame((sale),columns=["salary"])
    print(i, "          ", df3.salary.mean(), "          ", df3.salary.median())
    sale.clear()
    f = 0
```

```python
# Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?
b = df2[['job_title',"salary"]]

spisok = []
for i in b.job_title:
    if spisok.count(i) == 0:
        spisok.append(i)
    else:
        continue
sale = []
f = 0
maxx = 0
prof = ""
for i in spisok:
    for j in b.job_title:
        if i == j:
            sale.append(b.loc[f,'salary'])
            f += 1
        else:
            f += 1
    df3 = pd.DataFrame((sale),columns=["salary"])
    if maxx<df3.salary.mean():
        maxx = df3.salary.mean()
        prof = i
    sale.clear()
    f = 0
print(prof)
```

```python
# Какое процентное соотношение каждого региона по вакансиям от всех вакансий?
b = df2[['company_location']]

spisok = []
for i in b.company_location:
    if spisok.count(i) == 0:
        spisok.append(i)
    else:
        continue
count = 0
a = 0
for i in spisok:
    for j in b.company_location:
        if i == j:
            count += 1
    print(i," ",count*100/607)
    a += count*100/607
    count = 0
print(a)
```
